# **Section 1: Project Overview**

**"Fading Footprints"** is a project aimed to create and amplify social awareness around missing immigrants, their whereabouts and understanding the factors leading to the tragic fate many of these immigrants meet. With "Fading Footprints", the focus is on raising awareness about missing immigrants, shedding light on the factors surrounding their disappearances, and analyzing incidents involving migrants from 2014 to 2023. This project contains information about the date of report, location, source of the report, cause of death, gender and other crucial factors about these missing immigrants from various parts of the world. We will examine 13,000+ cases based on the aforementioned characteristics in order to confront the hidden reality of immigrant mortality, geopolitical ramifications, etc. The website Missing Migrants periodically updated with fresh cases of missing individuals as more and more data is produced since it is live data.

### **Shedding Light on Hidden Realities – Why is this important?**

Discovering the truths behind missing immigrants sheds light on a humanitarian concern often overlooked. By spotlighting these instances, the importance of safeguarding migrants' rights, lives, and acknowledging their vulnerability is emphasized.

### **Imputing**

**Imputing the 'Number of Dead' Column**

The "Number of Dead" column in our dataset represents immigrants who passed away during or after migration. During the data cleaning process, we noticed a substantial number of missing values in this column, creating uncertainty about the fate of those immigrants. To facilitate analysis, we've chosen to consider these individuals as missing rather than classifying them definitively as deceased. **Website Date** <br>
Here in our data we also have the data of the 'website date' which correlates to the the date on which the incident was reported and also the information was captured. For our dataset in the case of Website Date, we have seen the pattern that the records are in the order of incident reported date. Since there are only very few missing values in this column, we will backward fill with the next latest date of reporting. This will help us keep our dates synchronized with the incidents.

```python
# Imputing Website Date
# We'll be filling the website dates  with backward filling as the records are in the order of reported incidents
missing_migrants_data['Website Date'] = missing_migrants_data['Website Date'].fillna(method='bfill')
```

**Region of Origin** <br>

There are negligible amount of missing values in the 'Region of Origin' In the case of Country of Origin, there are a few records which have 'Unknown' along with other countries in the entry. We'll be removing the 'Unknown' part from the Country of Origin. There are negligible amount of missing values in the 'Region of Origin' column and many records with 'Unknown' entry. To address this, we employ a logic that infers the missing 'Region of Origin' based on the 'Country of Origin'. This approach is based on the assumption that there is a relation between the migrants country and region, of origin. We'll be taking the most frequently used region of origin while imputing the data based on the country.

```python
# Remove the Unknown part on Country of Origin where there is already a Country present
missing_migrants_data['Country of Origin'] = missing_migrants_data['Country of Origin'].apply(
    lambda x: x.split(',Unknown')[0] if pd.notna(x) else x
    )
missing_migrants_data['Country of Origin'] = missing_migrants_data['Country of Origin'].apply(
    lambda x: x.split('Unknown,')[1] if pd.notna(x) and len(x.split('Unknown,')) > 1 else x
)

# Create a dictionary mapping for existing Country of Origin to Region based on the most frequent region
country_to_region = missing_migrants_data.dropna(subset=['Region of Origin']).groupby('Country of Origin')['Region of Origin'].agg(pd.Series.mode).to_dict()
del country_to_region['Unknown']

# Function to fill missing 'Region of Origin' using the mapping
def impute_region_based_on_dictionary(row):
    if (pd.isna(row['Region of Origin']) or row['Region of Origin']=='Unknown') and row['Country of Origin'] in country_to_region:
        return country_to_region[row['Country of Origin']]
    else:
      return row['Region of Origin']

# Apply the impute function to the missing migrants data
missing_migrants_data['Region of Origin'] = missing_migrants_data.apply(impute_region_based_on_dictionary, axis=1)
```

**Minimum Estimated Number of Missing**

```python
# Finding common null values of number of dead and minimum estimated number of missing
common_null_values = missing_migrants_data['Number of Dead'].isna() & missing_migrants_data['Minimum Estimated Number of Missing'].isna()
common_null_values.value_counts()
```

Considering that no row contains both "Number of Dead" and "Minimum Estimated Number of Missing" values missing (neither are NaN simultaneously), we can infer that the "Total Number of Dead and Missing" column might be utilized to fill both the "missing" and "dead" columns.

**The equation utilized:**

**Total Number of Dead and Missing = Number of dead + Minimum estimated number of missing**

By subtracting the known values from the "total" figure, it is possible to derive the missing values in both columns since we have data for both "dead" and "missing" in each row.

```python
# Imputing Number of Dead & Minimum Estimated Number of Missing
missing_migrants_data['Number of Dead'].fillna(missing_migrants_data['Total Number of Dead and Missing'] - missing_migrants_data['Minimum Estimated Number of Missing'], inplace=True)
missing_migrants_data['Minimum Estimated Number of Missing'].fillna(missing_migrants_data['Total Number of Dead and Missing'] - missing_migrants_data['Number of Dead'], inplace=True)
```

There are now values in the "dead" and "missing" columns, and neither column has any missing data. This makes it possible to obtain accurate and full data for later analysis.

Here, we can observe that even though we did not have any missing values, the unknown values had already been classified as "Mixed or Unknown".  

The 'Cause of Death' column requires refinement with regards to its unique values. Looking more closely at the data, it is apparent that we can make it more readable by grouping the data points into distinct categories. The goal of this cleaning process is to enable more insightful observations and inferences from the data.

```python
# Defining a dictionary to map specific causes of death to their corresponding categories
category_cause_of_death = {
    'Drowning': 'Drowning',
    'Violence': 'Violence',
    'Accidental death': 'Accidental Death',
    'Vehicle accident / death linked to hazardous transport': 'Vehicle Accident',
    'Mixed or unknown': 'Mixed or Unknown',
    'Harsh environmental conditions / lack of adequate shelter, food, water': 'Harsh Environmental Conditions',
    'Sickness / lack of access to adequate healthcare': 'Lack of Access to Adequate Healthcare'}
```

We begin by classifying the entries into distinct categories according to our understanding of the data. Subsequently, we choose to assign these categories more significant names. As observed above, the last 8 unique values display a mixture of these categories with only a few occurrences. As such, we decide to include them under the general "Mixed or Unknown" category itself.

```python
# Function to categorize the cause of death using the dictionary
def categorize_cause_of_death(cause_of_death):

  category = category_cause_of_death.get(cause_of_death)
  if category is not None:
    return category
  else:
    return 'Mixed or Unknown'

# Apply the categorization function on 'Cause of Death' column
missing_migrants_data['Cause of Death'] = missing_migrants_data['Cause of Death'].apply(categorize_cause_of_death)
```

It is evident that the previous 8 unique values, with considerbaly lower frequency, have been clubbed with the "Mixed or Unknown" category.

The 'Cause of Death' column is hence, clean and prepared for data analysis.

# 2.Which are the popular information sources that cover news about the immigrants? Do they rely on information from multiple media reports?


```python
def clean_information_source(text):
    # Convert to lowercase
    text = text.lower()
    # Remove text within parentheses
    text = re.sub(r'\(.*?\)', '', text)
    # Correct specific spelling errors
    text = text.replace('initative', 'initiative')
    # Remove apostrophes
    text = text.replace("'", "")
    return text.strip()

missing_migrants_data['Information Source'] = missing_migrants_data['Information Source'].astype(str).apply(clean_information_source)
```

```python
# Extracting the relevant columns for the nested pie chart
source_quality = missing_migrants_data['Source Quality']
information_source = missing_migrants_data['Information Source']

# Grouping the data and getting the top information sources for each source quality
grouped_data = missing_migrants_data.groupby('Source Quality')['Information Source'].value_counts().groupby(level=0).head(3)

# Preparing the data for the nested pie chart
outer_values = grouped_data.values
outer_labels = [f'{info_src} ({count})' for info_src, count in grouped_data.items()]
inner_labels = grouped_data.index.get_level_values(0).unique()
inner_values = grouped_data.groupby(level=0).sum()

base_colors = ['#FF6B45', '#0093AF', '#32CD32', '#C21807', '#FFAB05']

# Function to generate shades for the information sources
def get_shades(base_color, num_shades):
    return [plt.cm.colors.to_rgba(base_color, alpha=i) for i in np.linspace(0.7, 0.3, num_shades)]

# Assigning colors to the outer ring
outer_colors = []
for i, source_quality in enumerate(inner_labels):
    num_shades = grouped_data.xs(source_quality).shape[0]
    base_color = base_colors[i % len(base_colors)]  # Cycling through the base colors
    outer_colors.extend(get_shades(base_color, num_shades))

# Creating the nested pie chart with increased size
fig, ax = plt.subplots(figsize=(15, 10))  # Increased figure size

# Outer ring with outlines
wedges, _ = ax.pie(outer_values, radius=1.3, colors=outer_colors, labeldistance=1.05,
                   wedgeprops=dict(edgecolor='white', linewidth=2))

# Inner ring with outlines and increased font size and bold font
_, autotexts = ax.pie(inner_values, radius=1, labels=inner_labels, labeldistance=0.7,
                      colors=base_colors[:len(inner_labels)], wedgeprops=dict(edgecolor='white', linewidth=2))
for autotext in autotexts:
    autotext.set_fontsize(12)  # Increase font size
    autotext.set_weight('bold')  # Make font bold

# Adding a legend for the outer ring labels with reduced font size
ax.legend(wedges, outer_labels, title="Information Sources", loc="center left", bbox_to_anchor=(1.05, 0, 0.5, 1), fontsize='small')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.set_aspect('equal')

plt.title("Distribution of Top Information Sources by Source Quality", fontsize=20, pad=30)

plt.show()
```
